```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
```

## Trial with yahoo Finance api 

```python
# tickers = ['AAPL', 'AXP', 'BA', 'CAT', 'CSCO', 'CVX', 'DIS', 'DOW', 'GS', 'HD']
tickers = ['AAPL']
start_time = datetime.strptime('2022-9-19','%Y-%m-%d')
end_time = datetime.strptime('2022-10-18','%Y-%m-%d')
for idx, company in enumerate(tickers):
    operator = yf.Ticker(company)
    share_outstanding = operator.info['sharesOutstanding']
    end = start_time + timedelta(days = 7)
    base = yf.download(tickers = tickers, start=start_time, end = end, interval="1m",auto_adjust = True,threads=True)
    while end < end_time:
        start_time += timedelta(days = 7)
        end += timedelta(days = 7)
        hist = yf.download(tickers = tickers, start = start_time, end = end, interval="1m",auto_adjust = True,threads=True)
        base = base.append(hist)
#     base = base.drop(['Dividends','Stock Splits'],axis = 1)
    base['Return'] = base['Close'].pct_change().fillna(0)
    base['free_turn'] = base['Volume'].div(share_outstanding)
    base.to_csv('{}.csv'.format(company))
    
```
